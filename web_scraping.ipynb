```python
import numpy as np
import time

titles = []
links = []

url = 'https://www.vogue.com/fashion/trends#fall%20fashion'


headers = random.choice(headers_list)
r = requests.Session()
r.headers = headers

time.sleep(5)
time.sleep(np.random.uniform(5,15))
html = r.get(url).text
            
soup = bs.BeautifulSoup(html, "html.parser")

infotable = soup.find("section", class_="FilterableSummaryListWrapper-kWopfW eAJhnu FilterableSummaryList-crZmYp iXhDyr")

for item in infotable.find_all("li", class_="CarouselListItem-likXzW"):
    link = item.find("div", class_="ClampContent-hilPkr eKRyDY")
    title = item.find("div", class_="ClampContent-hilPkr eKRyDY")
    
    links.append(link.a['href'])
    titles.append(title.getText())
    
```

```python
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service

chrome_options = Options()
#chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(service=Service(), options = chrome_options)
time.sleep(3)

titles = []
products = []

for url in links_fin:
    try:
        driver.get(url)
     
        element_present = EC.presence_of_element_located((By.CLASS_NAME, 'BasePageMainContent-kFvAuh'))
        WebDriverWait(driver, 10).until(element_present)
        
        SCROLL_PAUSE_TIME = 5
        last_height = driver.execute_script("return document.body.scrollHeight")
        
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
            
        time.sleep(random.uniform(15,25))
        
        html = driver.page_source
        soup = bs(html, 'html.parser')
        relevant = soup.find_all('main', class_="BasePageMainContent-kFvAuh ewPCZH page__main-content")
        
        for li in relevant:
            for title_class in li.find_all('h1', class_="BaseWrap-sc-gjQpdd BaseText-ewhhUZ ContentHeaderHed-NCyCC iUEiRd bwDymH isouMH"):
                t = title_class.getText()
                titles.append(t)
            for product_class in li.find_all('div', class_="BaseWrap-sc-gjQpdd BaseText-ewhhUZ UnifiedProductCardNameWrapper-fHyUwm iUEiRd fcKlBk jndIQb"):
                p = product_class.getText()
                products.append(p)
                
    except Exception as e:
        print(f"Error encountered for URL {url}: {e}")

driver.quit()
```
